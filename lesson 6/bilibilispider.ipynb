```python
import requests
```

```python
url = 'https://search.bilibili.com/all?keyword=%E4%B8%9C%E6%96%B9%E6%B0%B8%E5%A4%9C%E6%8A%84&from_source=banner_search&spm_id_from=333.334.b_62616e6e65725f6c696e6b.1'
res = requests.get(url)
res.encoding = 'utf-8'
```

```python
from bs4 import BeautifulSoup

text = res.text
soup = BeautifulSoup(text, 'html.parser')
```

```python
print(soup.prettify())
```

```python
titles = soup.find_all('a', class_ = 'title')
print(titles)
```

```python
for title in titles:
    print(title.text)
```

```python
for title in titles:
    print('https:' + title['href'])
```

```python
import re

pattern = re.compile('<a.*?video/(.*?).?from')
for title in titles:
    string = str(title)
    id = re.findall(pattern, string)[0]
    print(id)
    pass
```

```python
info = {}
pattern = re.compile('<a.*?video/(.*?).?from')
for title in titles:
    string = str(title)
    av_id = re.findall(pattern, string)[0]
    title_text = title.text
    av_url = 'https:' + title['href']
    info[av_id] = (title_text, av_url)
    pass

print(info)
```

```python
video = 'av811902'
title, url = info[video]
```

```python
print(title)
print(url)
```

```python
url = 'https://www.bilibili.com/video/av811902/?p=1'
res = requests.get(url, headers=headers, stream=True, verify=False)
res.encoding = 'utf-8'
pattern = '.__playinfo__=(.*)</script><script>window.__INITIAL_STATE__='
infos = re.findall(pattern, res.text)[0]
print(infos)
```

```python
import json
j = json.loads(infos)
print(j)
```

```python
durl = j['data']['durl']
print(durl)
```

```python
urls = [url['url'] for url in durl]
print(urls)
```

```python
for url in urls:
    print(url)
    pass
```

```python
import os

for index, url in enumerate(urls):
    session = requests.Session() 
    response = session.get(url, headers=headers, stream=True, verify=False)
    chunk_size = 1024 * 800
    content_size = int(response.headers['content-length'])
    video_path = os.path.join('./videos/av811902_p1.flv')
    if response.status_code==200:
        size = 0
        with open(video_path, 'ab') as file:
            for data in response.iter_content(chunk_size = chunk_size):
                file.write(data)
                size += len(data)
                file.flush()
                if size!=content_size:
                    print('第%d个片段：[下载进度]: %0.2fMB/%.2fMB' % (index+1, float(size/1024/1024), (content_size/1024/1024)), end='\r')
                else:
                    print('第%d个片段：[下载进度]: %0.2fMB/%.2fMB' % (index+1, float(size/1024/1024), (content_size/1024/1024)))
                pass
            print('[ok]')
        pass
    else:
        print('链接异常')   
        pass
                
```
